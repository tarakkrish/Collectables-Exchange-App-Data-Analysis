```python
#Create a copy of dataset "D6_data_renamed_columns" and assign to "df1" for the use of creating first graph
df1=D6_data_renamed_columns
```

```python
result1 = df1.groupby('How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?').size().reset_index(name='count of votes for prime membership')
```

```python
#Create a new row with value 0 and column name "How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?" and assign it to the variable "new_row".
new_row = pd.DataFrame({'How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?':0,'count of votes for prime membership':1},
                                                            index =[0])
#Use 'concat' to add the 'new_row' to 'result1'.
result1 = pd.concat([new_row, result1]).reset_index(drop = True)
# Use the "head()" function to show the first five rows of the content of 'result1' for data inspection.
result1.head()
```

```python
# Use 'groupby' to group the DataFrame by the values in the specified column, then use 'size' to count the occurrences of each group, and finally use 'reset_index' to convert the Series to a DataFrame and rename the count column and assign it to the variable 'result2'.
result2 = df1.groupby('How much would you be willing to spend per order using an app that allows you to buy and sell collectable items?').size().reset_index(name='count of votes willing to spend per order')
# Use the "head()" function to show the first five rows of the content of 'result2' for data inspection.
result2.head()
```

```python
# Use the "merge()" from a library called "pandas" that was imported as "pd" to merge two datasets that were assigned to the variables "result1" and "result2", respectively, on the common column called "df1result".
df1result = result1.merge(result2, left_on='How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?', right_on='How much would you be willing to spend per order using an app that allows you to buy and sell collectable items?')
# Use the "head()" function to show the first five rows of the content of 'result2' for data inspection.
df1result.head()
```

```python

# Prepare data for plotting.
#Get the value from 'count of votes for prime membership' and assign it to the variable 'votes_prime'.
votes_prime = df1result['count of votes for prime membership']
#Get the value from 'count of votes willing to spend per order' and assign it to the variable 'votes_order'.
votes_order = df1result['count of votes willing to spend per order']
labels = df1result['How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?']


fig, ax = plt.subplots(figsize=(10, 6))

# Create the stacked bar chart for Prime Membership.
bottom = 0
# Define a list of colors for the bars
colors = ['#FF5733', '#33FF57', '#3357FF', '#A333FF', '#FF8C33']
# Loop through each index in the range of the length of votes_prime
for i in range(len(votes_prime)):
    # Create a bar for 'Prime Membership' with the current vote count, stacking it on top of the previous bars
    ax.bar('Prime Membership', votes_prime[i], bottom=bottom, color=colors[i], label=f'Prime {labels[i]}' if i == 0 else "")
    # Increment the bottom position for the next bar by the current vote count
    bottom += votes_prime[i]

# Create the stacked bar chart for order spending.
bottom = 0
# Loop through each index in the range of the length of votes_order.
for i in range(len(votes_order)):
        # Create a bar for 'Prime Membership' with the current vote count, stacking it on top of the previous bars.
    ax.bar('Order Spending', votes_order[i], bottom=bottom, color=colors[i], label=f'Order {labels[i]}' if i == 0 else "")
    # Increment the bottom position for the next bar by the current vote count.
    bottom += votes_order[i]

# Customize plot title, xlabel, ylabel
plt.title('Votes for Prime Membership and Order Spending')
plt.xlabel('Spending Type')
plt.ylabel('Count of Votes')

# Create custom legends
legend_prime = [plt.Rectangle((0, 0), 1, 1, color=colors[i]) for i in range(len(labels))]
legend_order = [plt.Rectangle((0, 0), 1, 1, color=colors[i]) for i in range(len(labels))]
legend_label_prime = ['I would not use (0)','$4.99 or less (1)','$5 to 6.99 (2)','$7 to 8.99 (3)','$9 or more (4)']
legend_label_order = ['I would not use (0)','$4.99 or less (1)','$5 to 9.99 (2)','$10 to 14.99 (3)','$15 or more (4)']	

#Use legend to create a legend box with the legend_label_prime details and assign it to the variable first_legend.
first_legend = plt.legend(legend_prime, legend_label_prime, title='Prime Membership', loc='upper left', bbox_to_anchor=(1, 0.5))
#Use legend to create a legend box with the legend_label_order details and assign it to the variable second_legend.
second_legend = plt.legend(legend_order, legend_label_order, title='Order Spending', loc='upper left', bbox_to_anchor=(1, 0.8))

# Add legends to the plot.
ax.add_artist(first_legend)
ax.add_artist(second_legend)

# Adjust layout to fit legends.
plt.tight_layout()
plt.savefig("Stacked bar chart of Prime Membership and Order Spending Votes.png")

 # To show the plot on the screen.
plt.show()
```

```python
#Use Copy to create copy of dataframe 'D6_data_renamed_columns' and assign to a variable 'df2'.
df2 = D6_data_renamed_columns.copy()
#Use the list selection method to select only two columns: 'How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?' and 'How much would you be willing to spend per order using an app that allows you to buy and sell collectable items?'.
df2 = df2[['How much would you be willing to spend per order using an app that allows you to buy and sell collectable items?','How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?']]
#Use 'head' to check the changes on dataframe 'df2'.
df2
```

```python
#Create a new dictionary to change the column names.
new_column_names = {'How much would you be willing to spend per order using an app that allows you to buy and sell collectable items?':'Per Order','How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?':'Per Membership'}
#Use rename to change the column names with the dictionary 'new_column_names'.
df2.rename(columns=new_column_names,inplace=True)
# Use the "head()" function to show the first ten rows of the content of D4_data for data inspection.
df2.head(10)
```

```python
# Count of votes for the highest category in both categories
count_4_4 = len(df2[(df2['Per Order'] == 4) & (df2['Per Membership'] == 4)])

# Count of votes for highest in only one category and second highest in another category.
count_4_3_and_3_4 = len(df2[((df2['Per Order'] == 4) & (df2['Per Membership'] == 3)) |
                          ((df2['Per Order'] == 3) & (df2['Per Membership'] == 4))])

# Count of highest in one and less than 3 in another
count_4_less_than_3 = len(df2[(df2['Per Order'] == 4) & (df2['Per Membership'] < 3)])
count_less_than_3_4 = len(df2[(df2['Per Order'] < 3) & (df2['Per Membership'] == 4)])
```

```python
from matplotlib_venn import venn2, venn3

# Prepare the venn diagram
plt.figure(figsize=(8, 6))
venn2(subsets=(count_4_3_and_3_4, count_4_4, count_less_than_3_4 + count_4_less_than_3),set_colors=('skyblue', 'lightgreen','lightcoral'))

# Custom legend
colors = ['skyblue','#e6fffe', 'lightgreen']
labels = ['Votes of highest in only one category', 'Votes of highest in both categories', 'Highest in one & Less than 3 in another']
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]
plt.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.7, 1))

plt.title("Venn Diagram for Comparsion of Highest Votes Per Order and Per Membership")
plt.savefig('Venn Diagram.jpg')
plt.show()  # To show the plot on the screen.
```

```python
# Define variables for easy reference this is Q16 and Q19 for 'Data' sheet
app_price = "How much would you be willing to pay for an app that allows you to buy and sell collectable items?"
prime_price = "How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?"
```

```python
#we will get value counts for each variables defined above and we sort them by index
app_counts = df1[app_price].value_counts().sort_index()
prime_counts = df1[prime_price].value_counts().sort_index()
```

```python

# Define the questions
app_price = "How much would you be willing to pay for an app that allows you to buy and sell collectable items?"
prime_price = "How much would you be willing to spend per month for a prime membership to an app that allows you to buy and sell collectable items?"

# Get value counts
app_counts = df1[app_price].value_counts().sort_index()
prime_counts = df1[prime_price].value_counts().sort_index()

# Create the plot
fig, ax = plt.subplots(figsize=(14, 8))  # Increased figure size to accommodate the text box

# Plot bars for app price (on the left side)
app_bars = ax.barh(app_counts.index, -app_counts.values, align='center', color='skyblue', label='One-time App Price')

# Plot bars for prime membership (on the right side)
prime_bars = ax.barh(prime_counts.index, prime_counts.values, align='center', color='lightgreen', label='Monthly Prime Membership')

# Add value labels to the bars
for bars in [app_bars, prime_bars]:
    for bar in bars:
        width = bar.get_width()
        label_position = width / 2 if width >= 0 else width / 2
        ax.text(label_position, bar.get_y() + bar.get_height()/2, f'{abs(int(width))}', 
                ha='center', va='center')

# Set labels and title
ax.set_xlabel('Count')
ax.set_ylabel('Response Options')
ax.set_title('Comparison of App Price vs Prime Membership Price')

# Add legend
ax.legend()

# Add vertical line at x=0
ax.axvline(x=0, color='black', linestyle='-')

# Adjust x-axis ticks
max_value = max(app_counts.max(), prime_counts.max())
ax.set_xticks(range(-max_value, max_value+1, 5))
ax.set_xticklabels([str(abs(x)) for x in ax.get_xticks()])

# Add text box explaining response options
# Create separate legends for each price type
app_price_text = """One-time App Price Options:
0: I would not use
1: $0.99 or less
2: $1 to $1.99
3: $2 to $2.99
4: $3 to $3.99
5: $4 to $4.99
6: $5 or more"""

prime_price_text = """Monthly Prime Membership Options:
0: I would not use
1: $4.99 or less
2: $5 to $9.99
3: $10 to $14.99
4: $15 or more"""

plt.text(1.05, 0.7, app_price_text, transform=ax.transAxes, fontsize=9,
         verticalalignment='center', bbox=dict(boxstyle='round', facecolor='skyblue', alpha=0.1))

plt.text(1.05, 0.3, prime_price_text, transform=ax.transAxes, fontsize=9,
         verticalalignment='center', bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.1))

# Adjust layout to make room for the text box
plt.tight_layout()
plt.subplots_adjust(right=0.75)
plt.savefig("Comparison of App Price vs Prime Membership Price.jpg")
plt.show() # To show the plot on the screen.
```
