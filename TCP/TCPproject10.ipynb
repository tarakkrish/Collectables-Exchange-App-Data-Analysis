```python
# grouped = Question1.groupby(['How much would you be willing to pay for an app that allows you to buy and sell collectable items?', 'Gender']).size().unstack()
grouped = Question1.value_counts(['How much would you be willing to pay for an app that allows you to buy and sell collectable items?', 'Gender']).unstack().fillna(0)

grouped
```

```python
desired_order = ['I would not use', '$0.99 or less', '\$1 to \$1.99', '\$2 to \$2.99', '\$3 to \$3.99', '\$4 to \$4.99', '$5 or more',]

# Reindex the DataFrame to reorder the index
grouped = grouped.reindex(desired_order)
grouped
```

```python
# grouped = Question1.groupby(['How much would you be willing to pay for an app that allows you to buy and sell collectable items?', 'Gender']).size().unstack()

# Calculate the percentage of grand total
percentages = grouped.div(grouped.sum().sum()).multiply(100)

# Plot the data
plt.figure(figsize=(10, 6))
percentages.plot(kind='line', marker='o')

# Customize the plot
plt.title('How much the target customers are willing to pay to use the Collectables Exchange App\n varies by gender distribution')
plt.xlabel('Amount willing to pay by target customers')
plt.ylabel('Percentage of Total Disrtibution \nof Target Customers')
plt.legend(title='Gender')
plt.xticks(rotation=45)
# plt.xticks(ticks=['I would not use','$1.99 or less', '\$2 to \$3.99', '\$4 to \$5.99', '$6 or more'],labels = ['I would not use','$1.99 or less', '\$2 to \$3.99', '\$4 to \$5.99', '$6 or more'],fontsize=8,rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
"""The line graph reveals that the willingness of the female gender is greater than
that of the male to pay for using the collectable exchange app. Undoubtedly, 
the summation of female percentage is more than 50%, with a distribution of 
around 23.6%, 15.7%, and 11% from $1 to $3.99, which is complete supremacy 
over the male gender.It is clear that 10% of total target customers are not at 
all willing to pay for just using the Collectables Exchnage App."""
```

```python
# Group by 'Income (monthly)' and calculate the percentage of 1s for each column
count_df = Question3.groupby('Income (monthly)').sum()
count_df
```

```python
desired_order = ['$499 or less', '\$500 to \$999', '\$1,000 to \$1,499', '\$1,500 to \$1,999', '$2,000 or more']

# Reindex the DataFrame to reorder the index
count_df = count_df.reindex(desired_order)
```

```python



# Define the categories and colors for the plot
categories = [
    'Likelihood of using the Collectibles Exchange App to purchase: Toys',
    'Likelihood of using the Collectibles Exchange App to purchase: Stamps',
    'Likelihood of using the Collectibles Exchange App to purchase: Antique furniture',
    'Likelihood of using the Collectibles Exchange App to purchase: Comic books',
    'Likelihood of using the Collectibles Exchange App to purchase: Coins and currency'
]
colors = ['b', 'g', 'r', 'c', 'm']

# Create the figure and axes
fig, ax = plt.subplots(figsize=(14, 8))

# Set the x locations for the groups
x = np.arange(len(count_df.index))

# Set the width of the bars
width = 0.15

# Plot each category
for i, category in enumerate(categories):
    ax.bar(x + i * width, count_df[category], width, label=category, color=colors[i])

# Set the title and labels
ax.set_title('Interest Levels of Target Customers (Highly Likely and Likely) in Purchasing \nToys, Stamps, Antique Furniture, Comic Books, and Coins/Currency by Monthly Income', fontsize=16)
ax.set_xlabel('Distribution of frequencies by target customers monthly income', fontsize=14)
ax.set_ylabel('Count of target customers interested \nin purchasing the products', fontsize=14)
ax.set_xticks(x + width * (len(categories) - 1) / 2)
ax.set_xticklabels(count_df.index, rotation=45)

# Add a legend
ax.legend(title='Likelihood of using the \nCollectibles Exchange App to purchase:', labels=['Toys', 'Stamps', 'Antique furniture', 'Comic books', 'Coins and currency'],bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
"""The provided bat chart data visualization (DV) reveals that the Toys consistently 
show high interest across all income brackets, 
indicating a widespread appeal regardless of income. 
Suprisingly, the target customers with the lower income levels ($499 or less per month) 
are more interested in purchasing the products(Toys, Antique furniture, Coins and currency), than with any other income levels."""
```
