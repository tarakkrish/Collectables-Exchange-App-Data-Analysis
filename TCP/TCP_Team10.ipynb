```python
# Define the likelihood mapping
likelihood_mapping = {
    "I will not use": 0,
    "Highly unlikely": 1,
    "Unlikely": 2,
    "Unsure": 3,
    "Likely": 4,
    "Highly likely": 5
}

# Convert likelihood to numbers
def convert_to_numbers(df, column): # Define function to convert likelihood responses to numeric values
    return pd.to_numeric(df[column].map(likelihood_mapping), errors='coerce')
    # Use pd.to_numeric() with map() method to convert strings to numbers
```

```python
# Calculate average for each ethnicity
def calculate_average(df, column): # Define function to calculate average likelihood by ethnicity
    numeric_column = convert_to_numbers(df, column) #Use convert_to_numbers() to transform string responses to numeric values
    return numeric_column.groupby(df['Ethnicity']).mean() 
#Use groupby() and mean() methods to calculate average likelihood for each ethnic group
```

```python
#Process data for each day
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
results = {} #Initialize empty dictionary to store results for each day

for day in days:
    column_name = f'How likely are you to use the Collectibles Exchange App on: {day}' # Use f-string to generate column name for each day
    if column_name in Processed_df.columns: # Check if the generated column name exists in the DataFrame
        results[day] = calculate_average(Processed_df, column_name) # Use calculate_average() function to compute and store results for each day
        
    else:
        print(f"Column not found: {column_name}")
```

```python
# Define the categories and colors for the plot
categories = ['Likelihood of using the Collectibles Exchange App to purchase: Toys','Likelihood of using the Collectibles Exchange App to purchase: Stamps','Likelihood of using the Collectibles Exchange App to purchase: Antique furniture','Likelihood of using the Collectibles Exchange App to purchase: Comic books','Likelihood of using the Collectibles Exchange App to purchase: Coins and currency']
colors = ['b', 'g', 'r', 'c', 'm']

# Use plt.subplots() to create a figure and a set of subplots with a specified size, assigning it to "fig" and "ax".
fig, ax = plt.subplots(figsize=(14, 8))

# Use np.arange() to create an array of x locations for the groups based on the length of the DataFrame index, assigning it to "x".
x = np.arange(len(count_df.index))

# Set the width of the bars
width = 0.15

# Use a for loop to iterate over the categories and plot each category as a bar in the bar plot.
for i, category in enumerate(categories):
    # Use ax.bar() to create a bar plot for each category with specific x locations, width, label, and color.
    ax.bar(x + i * width, count_df[category], width, label=category, color=colors[i])

# Use ax.set_title() to set the title of the plot with a specific font size.
ax.set_title('Interest Levels of Target Customers (Highly Likely and Likely) in Purchasing \nToys, Stamps, Antique Furniture, Comic Books, and Coins/Currency by Monthly Income', fontsize=16)
# Use ax.set_xlabel() to set the x-axis label with a specific font size.
ax.set_xlabel('Distribution of frequencies by target customers monthly income', fontsize=14)
# Use ax.set_ylabel() to set the y-axis label with a specific font size.
ax.set_ylabel('Count of target customers interested \nin purchasing the products', fontsize=14)
# Use ax.set_xticks() to set the locations of x-ticks based on the width and number of categories.
ax.set_xticks(x + width * (len(categories) - 1) / 2)
# Use ax.set_xticklabels() to set the labels for x-ticks with rotation for better readability.
ax.set_xticklabels(count_df.index, rotation=45)
# Use ax.legend() to add a legend to the plot with a title, labels for each category, and a specified position.
ax.legend(title='Likelihood of using the \nCollectibles Exchange App to purchase:', labels=['Toys', 'Stamps', 'Antique furniture', 'Comic books', 'Coins and currency'],bbox_to_anchor=(1.05, 1), loc='upper left')
# Use plt.tight_layout() to automatically adjust subplot parameters to give specified padding.
plt.tight_layout()
plt.show() # Use the "show()" function of the "pyplot" module from the "matplotlib" library that was imported as "plt" to show the chart.
```
