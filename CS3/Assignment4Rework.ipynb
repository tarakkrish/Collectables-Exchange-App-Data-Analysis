```python
# In this example Use Case, Q13 is the "How many hours per day do you listen to the radio, including online?" variable.
# I need to know what the range of the response option numeric codes are for the Q13 variable, so:
response_options = [(D4_data["Q13"].max() - D4_data["Q13"].min())]
response_options # running this line gives me the answer in output below
```

```python
# In this case, Q13 is the "How many hours per day do you listen to the radio, including online?" variable.
# Use rcParams to set the default figure size for plots to 10 inches wide by 6 inches tall.
plt.rcParams["figure.figsize"] = [10, 6]
# Use rcParams to enable automatic layout adjustments for plots to prevent overlapping elements.
plt.rcParams["figure.autolayout"] = True
xticks = [0, 1, 2, 3, 4, 5, 6]  # I created this list from looking in the Data Dictionary to see what the response options numeric codes were for Q13.
# Use sns.histplot() to create a histogram of the "Q13" column from the DataFrame "D4_data", with bins based on the range of response options plus one, and set edge color to black.
ax = sns.histplot(D4_data["Q13"], bins=range(response_options_range + 1), ec="k")
# Use set() method to add a chart title, x-axis title, and y-axis title to the histogram.
ax.set(title="How frequently do you listen to radio (including online) per day?", xlabel="Viewing Frequency Levels", ylabel="Target Customers Count per Viewing Frequency Levels")  # To create a histogram using the histplot() function, and adding a chart title, x-axis title, and y-axis title.
# Use set_xticks() method to set x-tick positions manually using the "xticks" list.
ax.set_xticks(xticks)
# Use set_xticklabels() method to label the x-ticks with a list of frequency descriptions, rotated 15 degrees for better readability.
ax.set_xticklabels(["Never", "Less than 1 hour", "1 to almost 3 hours", "3 to almost 5 hours", "5 to almost 7 hours", "7 to almost 9 hours", "9 or more hours"], rotation=15)
# Use savefig() to insert a customized file name and save your DV as a stand-alone image asset. Change it to .jpg if you want a .jpg.
plt.savefig("Radio_Usage-Histogram-Python.png")
# Use show() to display the plot on the screen.
plt.show()
```

```python
#Create a copy of dataset "D4_data_renamed_columns" and assign to "df2" for the use of creating further analysis
df2 = D4_data_renamed_columns.copy()
```

```python
# Use the 'drop' to remove the column called 'ID'.
df2 = df2.drop(columns=['ID'])
```

```python
#Use 'columns' to get the existing columns in the data frame 'df2'.
df2.columns
```

```python
# Mulitple two columns by 7 to convert the existing per-day data to per-week.
df2['How many hours per day do you watch TV, including online?']=df2['How many hours per day do you watch TV, including online?']*7
df2['How many hours per day do you listen to the radio, including online?']=df2['How many hours per day do you listen to the radio, including online?'] * 7
```

```python
# Plotting boxplots with 4 colors and assign them to a variable called 'colors'
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']

# Create a boxplot from DataFrame "df2", with patches colored and returning a dictionary of boxplot components.
box = df2.boxplot(patch_artist=True, return_type='dict')
# Iterate through the 'boxes' in the boxplot and set the face color of each box using the colors from the list.
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
# Set the title of the plot with a font size of 16.
plt.title('How Frequently Target Customers Spend Their Time on Multiple Media', fontsize=16)
# Label the x-axis with a description and set the font size to 12.
plt.xlabel('How Target customers spent their time', fontsize=12)
# Label the y-axis with a description and set the font size to 12.
plt.ylabel('Target Customers Count per Viewing Frequency Levels', fontsize=12)
# Set x-axis ticks manually to show 4 points and label them with a list: ['TV', 'Radio', 'Magazines', 'Newspaper'].
plt.xticks(ticks=[1, 2, 3, 4], labels=['TV', 'Radio', 'Magazines', 'Newspaper'], rotation=45, ha='right')  # Rotate x-axis labels for better readability.
# Add a grid to the y-axis with dashed lines and set the transparency to 0.7.
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust the subplot parameters to give some padding.
plt.savefig("Hours_Spent_on_Media-Boxplot-Python.png") # Save the plot as a PNG file named "Hours_Spent_on_Media-Boxplot-Python.png".
plt.show() # Display the plot on the screen.
```
