```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2['Month'] = df2['Date'].dt.to_period('M')
```

```python
# Use the 'def' to create 'calculate_mode' function with 'series' as a parameter.
def calculate_mode(series):
    # use '.mode()' attribute with the 'series' and assign it to the variable 'mode_values'
    mode_values = series.mode()
    #Use .empty to check if it is null or not.
    if not mode_values.empty:
        #Return the mode_value if not null.
        return mode_values.iloc[0]
    else:
        #Handle the Null value with None.
        return None

# Group by 'Month' and calculate mean and mode for 'Age'
#use .groupby with the column Month and aggregate the columns with mean and calculate_mode. Reset the 
result = df2.groupby('Month').agg(
    Average_Age=('Age', 'mean'),
    Most_Repeated_Age=('Age', calculate_mode)
).reset_index()
```

```python
#Use the set_index and set 'Date' to index.
df2 = df2.set_index('Date')
```

```python
#use the "resample()" and resample it with Month 'M'.
monthly_data = df2.resample('M')
```

```python
#Import stats from scipy for specific calculations of mode().
from scipy import stats
```

```python

#Calculate the average age for each month using .mean() on the 'Age' column of the 'monthly_data'.
average_age = monthly_data['Age'].mean()

# Calculate the mode age for each month using .mode() over the aggregate functions on the 'Age' column of the 'monthly_data'.
mode_age = monthly_data['Age'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)


```

```python
#Assign D1_D2_data_renamed_columns to a new variable df3 and display first rows of df3 with head() function
df3 = D1_D2_data_renamed_columns
df3.head()
```

```python
#Create a new column 'Month-Year' to extract month and year and discard day
#.dt uses datetime accessor and use .to_period('M') to convert each into a Monthly frequency object
df3['Month-Year'] = df3['Date'].dt.to_period('M')
df3.head()
```

```python
#calculate monthly average income by what college does your major belong to?
#Using df3.groupby to combine month-year and college major by monthly and calculating mean with .mean() function and reset index of final dataframe.
major_avg_income = df3.groupby(['What college does your major belong to?', 'Month-Year'])['Income (monthly)'].mean().reset_index()
```

```python
#Display first 5 rows by using head() function 
major_avg_income.head()
```

```python
#mapping dictionary of College major to replace column values of college major in above major_avg_income dataframe.
#we are using pandas map function to replace int values of the college major column in dataframe.
college_mapping = {
    1: "Arts & Letters",
    2: "Business",
    3: "Education",
    4: "Engineering",
    5: "Health & Human Services",
    6: "Professional & Fine Arts",
    7: "Sciences",
    8: "Undergraduate Studies",
    9: "Graduate Studies"
}
major_avg_income["What college does your major belong to?"] = major_avg_income["What college does your major belong to?"].map(college_mapping)
major_avg_income
```

```python
# Creating new arts_letters_income dateframe that filter and contains only Arts & Letters data from college major average income
# Sorting new arts_letters_income  by date
#Converting values of 'Month-Year' column into Datetime object
arts_letters_income = major_avg_income[major_avg_income['What college does your major belong to?'] == 'Arts & Letters']
arts_letters_income = arts_letters_income.sort_values('Month-Year')
arts_letters_income['Month-Year'] = arts_letters_income['Month-Year'].dt.to_timestamp()
```

```python
# Create plot with customize figsize
#Creates a line plot with 'Month-Year' in X-axis and Average monthly income in Y-axis with 'o' circular markers and Income label
plt.figure(figsize=(16, 10))
plt.plot(arts_letters_income['Month-Year'], arts_letters_income['Income (monthly)'], marker='o', label='Income')

# Customize the plot
#Rotate x-axis tick with 45 degrees to aviod overlapping and we use alpha=0.7 that sets opacity 70% to see grids with less color
#Adding legend to the plot on upper left with 10 font size
plt.title('Monthly Average Income for Arts & Letters Major over the time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Monthly Income', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Data', loc='upper left', fontsize=10)

# Create a second legend for income categories
legend_labels = [
    '$499 or less (1)',
    '$500 to $999 (2)',
    '$1,000 to $1,499 (3)',
    '$1,500 to $1,999 (4)',
    '$2,000 or more (5)'
]

# import Line2D from matplotlib.lines to create custom line custom_line objects with gray color [0][0] creates a line of zero lenght.
#using custom_lines handles and labels for this legend. 
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='gray', lw=2) for _ in range(len(legend_labels))]
second_legend = plt.legend(custom_lines, legend_labels, title='Income Categories', 
                           loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)

# Add the second legend manually by get current axis(gca) and adds seconds legend.
#tight_layout() automatically adjust the padding between subplots
plt.gca().add_artist(second_legend)
plt.tight_layout()
plt.subplots_adjust(right=0.85)

# Show the plot and saving the figure
plt.show()
plt.savefig('output.png')
```
