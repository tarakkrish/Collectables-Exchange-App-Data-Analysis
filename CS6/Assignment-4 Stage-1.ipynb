```python
# Use the "read_excel()" from a library called "pandas" that was imported as "pd" to save the D1 "Data Dictionary" sheet as a variable called "D4_DD" for later use
D4_DD = pd.read_excel(D4, "Data Dictionary")
```

```python
#Get the "ResponseNO" column from "D4_DD" data to Create list with variable "l"
#Later use that list "l" to rename the "D4_Data" columns
l = D4_DD["ResponseNO"]
print(l)
```

```python
labels_xaxis = D4_DD.iloc[2].values.tolist()
labels_xaxis = labels_xaxis[2:]
for i in range(0,len(labels_xaxis)):
    labels_xaxis[i] = labels_xaxis[i][0:-4]
print(labels_xaxis)
type(labels_xaxis)
```

```python
# Use the "rename()" function to create a Python dictionary called "D1_D2_data_renamed_columns" that consists of key-value pairs where the key (left of the colon :) are the column codes labels, and the value (right of the colon :) are the explanation labels for the column names.
D4_data_renamed_columns = D4_data.rename(columns={"Q12": l[0], "Q13": l[1], "Q14": l[2], "Q15": l[3]})
D4_data_renamed_columns.head()
#
```

```python
# Use the "to_excel()" function on the "D4_data_renamed_columns" variable holding the dataframe to a separate XLSX file called "D4_data_renamed_columns.xlsx" for later use.
# After this code executes, you will see the new XLSX file show up in the content pane to the left.
D4_data_renamed_columns.to_excel("/drive/cs6/D4_data_renamed_columns.xlsx")
```

```python
#Create a copy of dataset "D4_data_renamed_columns" and assign to "df1" for the use of creating first graph
df1=D4_data_renamed_columns
```

```python
df1.columns # To output a list of the column names in the dataset that was assigned to a variable called "df".
```

```python
# Programming a histogram in Python using a CSV file - histplot function
plt.ticklabel_format(style = 'plain') # To prevent scientific notation showing up in the Data Visualization (DV).
sns.histplot(df1["How many hours per week do you read newspapers, including online?"]).set(title="Total Newspaper Readers over hours", xlabel="", ylabel="Customers Count") # To create a histogram using the histplot() function, and adding a chart title, x-axis title, and y-axis title.
# plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}')) # To insert commas in the numbers on the x-axis tick labels.
# Set x-axis ticks manually to show 9 points
plt.xticks(ticks=range(0,8,1), labels=labels_xaxis)
plt.xticks(rotation=65)
plt.savefig("How many hours per week do you read newpapers, including online?.png") # To save the DV to a file. Change it to .jpg is you want a .jpg
plt.show() # To show the plot on the screen.
```
