```python
# Import required libraries
import pandas as pd # For data processing
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For plotting
import matplotlib as mpl # For plot formatting
```

```python
df = pd.read_csv('Class_Session_6_Activity.csv') # To load a CSV dataset file with a specific filename and assign it to a variable called "df".
```

```python
df.head() # To show the first five rows of the dataset that was assigned to a variable called "df".
```

```python
df.columns # To output a list of the column names in the dataset that was assigned to a variable called "df".
```

```python
# Programming a histogram in Python using a CSV file - histplot function
plt.ticklabel_format(style = 'plain') # To prevent scientific notation showing up in the Data Visualization (DV).
sns.histplot(df["Total_Trans_Amt"]).set(title="Total Transaction Amounts per Customer", xlabel="U.S. Dollars", ylabel="Customer Count") # To create a histogram using the histplot() function, and adding a chart title, x-axis title, and y-axis title.
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}')) # To insert commas in the numbers on the x-axis tick labels.
plt.savefig("BankChurner-Histogram-Python.png") # To save the DV to a file. Change it to .jpg is you want a .jpg
plt.show() # To show the plot on the screen.
```

```python
# Programming a boxplot in Python using an CSV file
sns.boxplot(x="Education_Level", y="Total_Revolving_Bal", data=df, palette="Set3").set(title="Revolving Balance by Customer Education Level", xlabel="", ylabel="U.S. Dollars") # To create a boxplot using the boxplot() function, adding a chart title, removing the default x-axis title because it is not needed in this case, and adding a y-axis title.
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}')) # To insert commas in the number on the y-axis tick labels.
plt.xticks(rotation=45) # For rotating the x-axis ticks 45 degrees so that the text doesn't overlap.
plt.savefig("BankChurner-Boxplot-Python.png") # To save the DV to a file. Change it to .jpg is you want a .jpg
plt.show() # To show the plot on the screen.
```

```python
# Programming a bubble chart in Python using an CSV file - build based on scatterplot function
sns.scatterplot(data=df, x="Total_Trans_Ct", y="Total_Trans_Amt", size="Customer_Age", hue="Customer_Age", legend="brief", ).set(title="Transaction Count and Amount Based on Customer Age", xlabel="Transaction Count", ylabel="Total Transaction Amount ($US)")  # To create a bubble chart built based on the scatterplot() function wherein variables x, y, bubble size, and bubble color (hue) are defined as specific columns from the dataset, specifying the legend type, adding a chart title, optimized x-axis title, and optimized y-axis title.
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}')) # To insert commas in the number on the y-axis tick labels.
plt.legend(title="Age", loc="best") # To optimize the legend name and position.
plt.savefig("BankChurner-Bubble_Chart-Python.png") # To save the DV to a file. Change it to .jpg is you want a .jpg
plt.show() # To show the plot on the screen.
# Optional: Teams, experiment with variations of the legend in the scatterplot() function: https://seaborn.pydata.org/generated/seaborn.scatterplot.html
```
