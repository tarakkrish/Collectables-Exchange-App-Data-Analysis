```python
# Create a basic line chart that shows change over time in python using a CSV file.
# Dataset source: https://serc.carleton.edu/hydromodules/steps/td1.html
# The data used in this Activity are Streamflow in cubic feet per second
```

```python
# Import basic libraries (A non-optimized code documentation line.)
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load a CSV file (A non-optimized code documentation line.)
df1 = pd.read_csv("/drive/Class_Session_4_Activity.csv")
```

```python
df1.head()
```

```python
print(df1)# Use the "print()" function to show the first and last five rows of a dataset that was assigned to a variable called "df1".
```

```python
df1.columns
```

```python
df1.describe()
```

```python
# Define a variable based on a column in a dataset (A non-optimized codedocumentation line.)
x1 = df1['Date']
x1
```

```python
# Define a variable based on a column in a dataset (A non-optimized codedocumentation line.)
y1 = df1['Streamflow']
y1
```

```python
plt.plot(x1, y1)
plt.xticks(x1[::365], rotation=25) # To make the x axis include all the dates that we have, and to rotate them to fit.
plt.ylabel("Streamflow (cfs)") # use the "ylabel()" function from the pyplot module of the matplotlib library to customize the y-axis label as "Streamflow (cfs)“where "cfs” is one way to abbreviate "cubic feet per second"
plt.title("Streamflow Over Time")
plt.savefig("Streamflow-CSV-Python.png") # To save the Python DV to a file. Change it to .jpg if you want a .jpg
plt.plot
```
